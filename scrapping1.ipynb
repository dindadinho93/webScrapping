```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://realpython.github.io/fake-jobs/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

```python
results = soup.find(id="ResultsContainer")
print(results.prettify())
```

```python
job_elements = results.find_all("div", class_="card-content")
```

```python
titles = []
companies = []
locations = []
for job_element in job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    titles.append(title_element.text.strip())
    companies.append(company_element.text.strip())
    locations.append(location_element.text.strip())
    print(title_element.text.strip())
    print(company_element.text.strip())
    print(location_element.text.strip())
    print()
```

```python
df = pd.DataFrame({'Title':titles,'Company':companies,'Location':locations})
df
```

```python
page = requests.get('https://news.detik.com/berita?tag_from=wp_news_firstnav_Berita')
soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find('div', class_='list-content')
print(results.prettify())
```

```python
news_elements = results.find_all('article', class_= 'ph_newsfeed_d')
news_elements
```

```python
title = []
link = []
img = []
for news_element in news_elements:
    title_element = news_element.find('a', class_='ai_link media__link')
    link_element = news_element['i-link']
    img_element = news_element['i-img']
    print(title_element['dtr-ttl'])
    print(link_element)
    print(img_element)
    print('###########################################')
    title.append(title_element['dtr-ttl'])
    link.append(link_element)
    img.append(img_element)
    
```

```python
df1 = pd.DataFrame({'Title': title, 'Link': link, 'Image': img})
```

```python
url = link[1]
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
results =soup.find('div', class_='page__breadcrumb')
results1 = soup.find('article', class_='detail')
print(results.prettify())
```

```python
kategori = []
title = []
author = []
date = []
kategori_elements = results.find('a', attrs={'dtr-sec':"breadcrumbkanal"})
kategori.append(kategori_elements.text.strip())
print(kategori)
header_elements = results1.find_all('div', class_='detail__header')
for header_element in header_elements:
    title_element = header_element.find('h1', class_="detail__title").text.strip()
    author_element = header_element.find('div', class_="detail__author").text.strip()
    date_element = header_element.find('div', class_="detail__date").text.strip()
    title.append(title_element)
    author.append(author_element)
    date.append(date_element)
    print(title)
    print(author)
    print(date)
```

```python
place = []
place_element = results1.find('strong').text.strip()
place.append(place_element)
print(place)
detail_elements = results1.find_all('p')
```

```python
content = []
cont_element = ''
for detail_element in detail_elements:
    content.append(detail_element.text.strip())
    cont = ''.join(content)
    content = [cont]
print(len(content))
```

```python
df2 = pd.DataFrame({'Link': url, 'Kategori': kategori, 'Title': title, 'Author': author, 'Date': date, 'Place': place, 'Content': content[-1]})
```

```python
df_merge_col = pd.merge(df1, df2, on=['Link', 'Title'], how='left')

df_merge_col
```
